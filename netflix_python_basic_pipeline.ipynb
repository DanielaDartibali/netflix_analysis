```python
import pandas as pd 
import numpy as np

df_netflix = pd.read_csv(r"netflix_titles_nov_2019.csv.zip")
df_netflix
```

```python
print(df_netflix.info())
print(df_netflix.isnull().sum())
```

```python
df_netflix.fillna({'director': 'Desconhecido', 'country': 'Desconhecido', 'rating': 'Sem classificação', 'date_added': df_netflix['date_added'].mode()[0]}, inplace=True)
df_netflix
```

```python
df_netflix['date_added']= pd.to_datetime(df_netflix['date_added'])
df_netflix['year_added'] = df_netflix['date_added'].dt.year
df_netflix['month_added'] = df_netflix['date_added'].dt.month
df_netflix 
```

```python
df_netflix['listed_in'] = df_netflix['listed_in'].str.split(', ')
df_netflix
```

```python
df_netflix['new_content'] = df_netflix['year_added'].apply(lambda x: 'Yes' if x >= 2019 else 'No')
df_netflix
```

```python
def classify_rating(rating):
    if rating in ['G', 'TV-Y', 'TV-G']:
        return 'Kids'
    elif rating in ['PG', 'TV-Y7', 'TV-PG']:
        return 'Family'
    elif rating in ['PG-13', 'TV-14']:
        return 'Teen'
    elif rating in ['R', 'TV-MA', 'NC-17']:
        return 'Adult'
    else:
        return 'Unknown'  

df_netflix['rating'] = df_netflix['rating'].apply(classify_rating)

print(df_netflix[['rating']].head())
```
